```python
blur1 = cv2.blur(img, (3, 3))
blur2 = cv2.blur(img, (9, 9))
blur3 = cv2.blur(img, (125, 125))
```

```python
fig=plt.figure(dpi=150)
plt.subplot(221),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(222),plt.imshow(blur1),plt.title('Blurred 3x3')
plt.xticks([]), plt.yticks([])
plt.subplot(223),plt.imshow(blur2),plt.title('Blurred 9x9')
plt.xticks([]), plt.yticks([])
plt.subplot(224),plt.imshow(blur3),plt.title('Blurred 125x125')
plt.xticks([]), plt.yticks([])
```

```python
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
```

```python
blur = cv2.blur(img, (9, 9))
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred Again')
plt.xticks([]), plt.yticks([])
```

```python
kernel12 = np.ones((12, 12), np.float32) / 144
image = cv2.filter2D(img, -1, kernel12)

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(image),plt.title('Kernel Blurred')
plt.xticks([]), plt.yticks([])
```

```python
kernel12 = np.ones((5, 5), np.float32) / 25
image2 = cv2.filter2D(img, -1, kernel12)

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(image2),plt.title('Kernel Blurred 2')
plt.xticks([]), plt.yticks([])
```
