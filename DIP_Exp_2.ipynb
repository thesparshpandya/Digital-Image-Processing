```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from google.colab import files

uploaded = files.upload()
```

```python
img = cv2.imread('Chess.jpg',0)
plt.imshow(img, cmap='gray')
```

```python
hist_values = cv2.calcHist([img], channels=[0], mask=None, histSize=[256], ranges=[0,256])
plt.plot(hist_values)
```

```python
equ = cv2.equalizeHist(img)
plt.imshow(equ, cmap='gray')
```

```python
hist_value = cv2.calcHist([equ], channels=[0], mask=None, histSize=[256], ranges=[0,256])
plt.plot(hist_value)
```

```python
res = np.hstack((img, equ))
plt.imshow(res, cmap='gray')
```

```python
plt.subplot(221),plt.imshow(img,cmap='gray'),plt.title('Original')
plt.subplot(222),plt.imshow(equ,cmap='gray'),plt.title('Equalized')
plt.subplot(223),plt.plot(hist_values),plt.title('Histogram')
plt.subplot(224),plt.plot(hist_value),plt.title('Histogram Equalized')
```
