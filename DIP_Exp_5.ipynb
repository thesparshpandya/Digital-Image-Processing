```python
# DFT
import numpy as np
import cv2
import matplotlib.pyplot as plt
from google.colab import files

uploaded = files.upload()

img = cv2.imread('einstein.jpg', 0)
imf = np.float32(img) / 255.0

dft = cv2.dft(imf, flags=cv2.DFT_COMPLEX_OUTPUT)

idft = cv2.idft(dft)
rec_img = cv2.magnitude(idft[:, :, 0], idft[:, :, 1])
rec_img1 = rec_img * 255.0
rec_img2 = np.uint8(rec_img1)

plt.subplot(121), plt.imshow(img, cmap='gray')
plt.subplot(122), plt.imshow(rec_img2, cmap='gray')
plt.show()

print(rec_img2)
print(img)

error = img - rec_img2
print(error)

SE = error * error
MSE = np.sum(SE)
print(MSE)

MSE1 = MSE / img.shape
print(MSE1)
```
